**symboling**: el grado de riesgo asociado con el seguro del automóvil, un valor más alto es un mayor riesgo  
**CarName**: la marca del coche o modelo  
**fueltype**: el tipo de combustible   
**aspiration**: el tipo de aspiración del motor  
**doornumber**: el número de puertas del coche  
**carbody**: el tipo de carrocería  
**drivewheel**: el tipo de tracción del automóvil (delantera, trasera o cuatro ruedas)  
**enginelocation**: la ubicación del motor en el vehículo (front o rear)  
**wheelbase**: la distancia entre los ejes delanteros y traseros   
**carlength**: la longitud total del coche  
**carwidth**: el ancho total del coche  
**curbweight**: el peso del coche  
**enginesize**: el tamaño del motor  
**fuelsystem**: el sistema de alimentación de combustible (inyección electrónica de combustible o carburador)  
**boreratio**: el diámetro de los cilindros del motor  
**stroke**: la distancia recorrida por el pistón en un solo movimiento dentro del cilindro del motor  
**compressionratio**: la relación de compresión del motor, que es la proporción entre el volumen del cilindro cuando el pistón está en el punto más bajo y cuando está en el punto más alto  
**horsepower**: la potencia del motor  
**peakrpm**: las revoluciones por minuto (rpm) a las que el motor alcanza su máxima potencia  
**citympg**: el rendimiento de combustible en conducción normal  
**highwaympg**: el rendimiento de combustible en carretera  
**price**: el precio del coche en dólares
